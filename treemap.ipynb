```python
def produce_prls(str_name):
    pr_ls = str_name.split(', ')
    return pr_ls

def find_grey(pr_ls, quali_ls):
    grey = {}
    ln_ls = []
    fn_ls = []
    for j in quali_ls:
        tmp = j.split()
        ln = tmp[-1]
        fn = tmp[0]
        ln_ls.append(ln)
        fn_ls.append(fn)
    for i in ln_ls:
        ind = ln_ls.index(i)
        if (i in pr_ls) or (quali_ls[ind] in pr_ls) or (fn_ls[ind] in pr_ls):
            pass
        else:
            grey[quali_ls[ind]] = ind
    print(grey)
    return grey

def gen_ax_df(cat, ateam, df):
    '''
    returnlist: [size, label, value<xg xg/90>, last name]
    '''
    a = df[(df['Squad'] == ateam) & (df['Pos'] != 'GK')]
    a = a.sort_values(by=['Min'], ascending=False, na_position='last')
    valid = a[:18]
    valid_cat = valid.sort_values(by=[cat], ascending=False, na_position='last')
    namel = valid_cat['Player'].tolist()
    xgl = valid_cat[cat].tolist()
    xgpers = cat + '90'
    xgperl = valid_cat[xgpers].tolist()
    rvalue = []
    for i in range(len(xgl)):
        tmp = str(xgl[i]) + ' / ' + str(round(xgperl[i], 2))
        rvalue.append(tmp)
    lsname = []
    for j in namel:
        tmp = j.split()
        lsname.append(tmp[-1])
    rl = [xgl, namel, rvalue, lsname]
    print(ateam, cat, 'top 18: ', namel)
    return rl

def plot_ax(zh, cat, x):
    t1 = zh + '外场球员在场时的' + cat + '\n'
    plt.title(t1, fontsize=12)
    t2 = "\n   越" + x + "越好"
    plt.title(t2, loc='right', fontsize=10)
    plt.axis('off')
    plt.tick_params(top='off', right='off')
```

```python
def treemap(ateam, bteam, astr, bstr, azh, bzh):
    pra = produce_prls(astr)
    prb = produce_prls(bstr)
    df = pd.read_excel('04_09_2021.xlsx')  #manually input name

    #A team data
    a_data_xg = gen_ax_df('onxG', ateam, df)
    a_data_xga = gen_ax_df('onxGA', ateam, df)
    print(ateam, 'xg grey:')
    a_xg_grey = find_grey(pra, a_data_xg[1])  # print grey name list
    print(ateam, 'xga grey:')
    a_xga_grey = find_grey(pra, a_data_xga[1])
    print('----------------------------------------------------------------')

    #B team data
    b_data_xg = gen_ax_df('onxG', bteam, df)
    b_data_xga = gen_ax_df('onxGA', bteam, df)
    print(bteam, 'xg grey:')
    b_xg_grey = find_grey(prb, b_data_xg[1])  # print grey name list
    print(bteam, 'xga grey:')
    b_xga_grey = find_grey(prb, b_data_xga[1])

    #plotting...
    fig = plt.figure(figsize=(5.607, 12.131))
    ax = fig.add_subplot(411)
    plt.rcParams['savefig.dpi'] = 458

    a_data_xg[3][-1] = ''
    a_data_xg[2][-1] = ''
    cl1 = ["#FFCFE2"]*18
    for i in a_xg_grey:
        cl1[a_xg_grey[i]] = 'grey'
    #cl1[x] = '#FFCFE2'
    plot = squarify.plot(sizes=a_data_xg[0], label=a_data_xg[3], text_kwargs={'fontsize': 6, 'fontname': "arial"}, color=cl1,
                         value=a_data_xg[2], alpha=.8, edgecolor='white')
    plot_ax(azh, 'xG', '高')

    ax2 = fig.add_subplot(412)
    a_data_xga[3][-1] = ''
    a_data_xga[2][-1] = ''
    #a_data_xga[2][-2] = '1.39'
    #a_data_xga[2][-3] = '0.73'
    cl2 = ["#FFCFE2"]*18
    for i in a_xga_grey:
        cl2[a_xga_grey[i]] = 'grey'
    #cl2[x] = '#FFCFE2'
    plot = squarify.plot(sizes=a_data_xga[0], label=a_data_xga[3], text_kwargs={'fontsize': 6, 'fontname': "arial"}, color=cl2,
                         value=a_data_xga[2], alpha=.8, edgecolor='white')
    plot_ax(azh, 'xGA', '低')

    ax3 = fig.add_subplot(413)
    b_data_xg[3][-1] = ''
    b_data_xg[2][-1] = ''
    cl3 = ["#00DCB5"]*18
    for i in b_xg_grey:
        cl3[b_xg_grey[i]] = 'grey'
    #cl3[x] = '#00DCB5'
    #b_data_xga[3][x] = 'Raphinha'
    plot = squarify.plot(sizes=b_data_xg[0], label=b_data_xg[3], text_kwargs={'fontsize': 6, 'fontname': "arial"}, color=cl3,
                         value=b_data_xg[2], alpha=.8, edgecolor='white')
    plot_ax(bzh, 'xG', '高')

    ax4 = fig.add_subplot(414)
    b_data_xga[3][-1] = ''
    b_data_xga[2][-1] = ''
    cl4 = ["#00DCB5"]*18
    for i in b_xga_grey:
        cl4[b_xga_grey[i]] = 'grey'
    #cl4[x] = '#00DCB5'
    #b_data_xga[3][x] = 'Raphinha'
    plot = squarify.plot(sizes=b_data_xga[0], label=b_data_xga[3], text_kwargs={'fontsize': 6, 'fontname': "arial"},
                         color=cl4, value=b_data_xga[2], alpha=.8, edgecolor='white')
    plot_ax(bzh, 'xGA', '低')

    plt.axis('off')
    plt.tick_params(top='off', right='off')
    plt.gcf().text(0.04, 0.961, '总值/每90分钟', fontsize = 10)
    fig.tight_layout()
    plt.savefig(ateam + bteam + '1111.png')
    plt.show()
```
